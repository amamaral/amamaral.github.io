## Atenção

- O deslocamento tem natureza vetorial e $\Delta x$ pode ser negativo.
- <div class="fragment fade-up"> Seu módulo, $|\Delta x|$ é sempre positivo, mas</div>
- <div class="fragment fade-up"> Não confundir $|\Delta x|$ com a distância percorrida!</div>

### Exemplo:
Uma partícula sofre um deslocamento $3,4\text{ m}$ e depois recua os mesmos $3,4\text{ m}$. 

(a) Qual a distância percorrida? <span class="fragment fade-up">$3,4 + 3,4 = 6,8\text{ m}$.</span>

(b) Qual o módulo do deslocamento? <span class="fragment fade-up">$|3,4 - 3,4| = 0\text{ m}$.</span>

## Velocidade média $(\bar{v})$

- Quão rápido uma partícula se move?
- <div class="fragment"> Vel. média num intervalo $[t_1, t_2]$: </div>
<div class="fragment"> $$ \bar{v} = \frac{x(t_2) - x(t_1)}{t_2 - t_1} = \frac{x_2 - x_1}{t_2 - t_1}$$</div>
<div class="fragment"> No SI, $\bar{v}$ tem unidades m/s.</div>
- <div class="fragment"> Importante: independe dos detalhes do movimento. </div>

```python
# Coordenadas x e y dos pontos como "arrays"

x = array([2, 8])
y = array([2, 8])

# Representação gráfica dos pontos (x,y) acima
plot(x,y,'ro')

xlabel("t")
ylabel("x")
xticks(arange(0,11,2),['','$t_1$','','','$t_2$'])
yticks(arange(0,11,2),['','$x_1$','','','$x_2$'])
grid()
```

## Exemplo

Calcule $\bar{v}$ no intervalo $[1\text{ s},4\text{ s}]$, quando $x(t)=5t^2-t^3$.

$x(1\text{ s}) = 4$ m, e $x(4\text{ s}) = 16$ m.

$\bar{v} =$ <span class="fragment"> $\frac{\Delta x}{\Delta t} = $ </span> <span class="fragment"> $ \frac{x(4)-x(1)}{4-1} = $ </span> <span class="fragment"> $ \frac{12}{3} = 4$ </span>

<div class="fragment"> $\bar{v}$ é associado com a inclinação da secante. ver <font color="red">WolframAlpha.com</font>.</div>

```python
# Coordenadas x e y dos pontos como "arrays"

t = linspace(0,5,100)

x = 5*t**2 - t**3

# Representação da função
plot(t,x)

plot([1,4], [4, 16], 'ko') # Pontos em t=1s e 4s
plot([0,5], [0, 20])       # Secante
plot([1,4], [4, 4],  '--') # Delta t
plot([4,4], [4, 16], '--') # Delta x

text(3.5, 10, '$\Delta x$')
text(2.5, 2,  '$\Delta t$')

xlabel("t (s)")
ylabel("x (m)")
grid()
```

## Velocidade instantânea $v$

- Velocidade relevante p. ex. em acidentes!
- <div class="fragment"> Como obter $v$? </div>
- <div class="fragment"> R.: Usando intervalos de tempo $(\Delta t)$ cada vez menores.</div>

#### Exemplo

Considere uma bola em queda livre, tal que $x(t)=\frac{gt^2}{2}\approx5t^2$, e que $t_2 = t_1 + \Delta t$, onde $t_1 = 1 $ s.

```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

def plot_secant(x):
    dt = x
    # Coordenadas x e y dos pontos como "arrays"

    t = linspace(0,5,100)

    x = 5*t**2

    # Representação da função
    plot(t,x)

    # Representação da secante
    t1 = 1
    t2 = t1 + dt
    x1 = 5
    x2 = 5*t2**2
    
    dx = x2-x1

    a = dx/dt
    b = 5 - a

    plot([t1,t2], [x1, x2], 'ko')  # Pontos em t=1s e 4s
    plot([0,5], [b, a*5+b])        # Secante
    plot([t1,t2], [x1, x1], '--')  # Delta t
    plot([t2,t2], [x1, x2], '--') # Delta x
    
    text(0.0, 60, """    x2-x1: %.2f m
    t2-t1: %.2f (s)
    v=%.2f m/s""" % (dx, dt, dx/dt))

    xlabel("t (s)")
    ylabel("x (m)")
    ylim(-5,100)
    grid()

interact(plot_secant, x=widgets.FloatSlider(min=0.01,max=4.0,step=0.01,value=3, description=r'$\Delta t$'));
```

## Velocidade instantânea $v$ (definição)

$$v = \lim_{\Delta t \rightarrow 0}\frac{x(t+\Delta t)-x(t)}{\Delta t} = \frac{dx}{dt}$$
- <div class="fragment"> $v=\frac{dx}{dt}$ é a "derivada de $x$ em relação a $t$".</div>
- <div class="fragment"> Mas, $x$ é função de $t$: derivada é a taxa de variação da posição.</div>
- <div class="fragment"> Unidade: $v=\frac{dx}{dt} \rightarrow \frac{\text{m}}{\text{s}}$ (no SI).</div>

Inicialmente ele está parado no intervalo [0 s, 1 s],<span class="fragment"> e que acelera uniformemente entre 1 s e  3 s.</span> <span class="fragment"> Sabe-se que $x(3) = 4$ m.</span> <span class="fragment"> Entre 3 s e 8 s, o elevador se move a uma velocidade uniforme, atingindo $x(8) = 24$ m,</span> <span class="fragment"> e desacelerando até o repouso total em $t=9$ s.

## Intuição geométrica para $v$

Represente a velocidade de um certo elevador que realiza o seguinte movimento. Inicialmente ele está parado no intervalo [0 s, 1 s], e que acelera uniformemente entre 1 s e  3 s. Sabe-se que $x(3) = 4$ m. Entre 3 s e 8 s, o elevador se move a uma velocidade uniforme, atingindo $x(8) = 24$ m, e desacelerando até o repouso total em $t=9$ s.

```python
t = linspace(0, 10, 500)
dt = t[1]-t[0]

a = 2*(greater_equal(t,1)*less_equal(t, 3))-4*(greater_equal(t,8)*less_equal(t, 9))

vi = 0
v = vi + a.cumsum()*dt # The last term is equivalent to the time integral of a

xi = 0
x = xi + v.cumsum()*dt # The last term is equivalent to the time integral of v

subplot(211)
plot(t, x)
grid()
ylim(-1, 30)
yticks(arange(6)*5)
xticks(arange(11))
ylabel('x(m)')

subplot(212)
plot(t, v)
grid()
ylim(-0.1,5)
yticks(arange(6))
xticks(arange(11))
ylabel('v(m/s)')
xlabel('t(s)')

tight_layout()
```

## Cálculo analítico de $v$

- Retornemos ao exemplo da bola em queda livre. $x(t)=5t^2$. Calcular $v$ em $t=1$ s.

<span class="fragment"> $v = \lim_{\Delta t \rightarrow 0}\frac{x(t+\Delta t)-x(t)}{\Delta t} =$ </span><span class="fragment"> $\lim_{\Delta t \rightarrow 0}\frac{5(1+\Delta t)^2-5(1)}{\Delta t}=$ </span><span class="fragment"> $\lim_{\Delta t \rightarrow 0}\frac{5(1+2\Delta t+\Delta t^2)-5}{\Delta t}=$</span><span class="fragment"> $\lim_{\Delta t \rightarrow 0}(10+5\Delta t)=10\frac{m}{s}$</span>

```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

def plot_secant(x):
    dt = x
    # Coordenadas x e y dos pontos como "arrays"

    t = linspace(0,5,100)

    x = 5*t**2

    # Representação da função
    plot(t,x)

    # Representação da secante
    t1 = 1
    t2 = t1 + dt
    x1 = 5
    x2 = 5*t2**2
    
    dx = x2-x1

    a = dx/dt
    b = 5 - a

    plot([t1,t2], [x1, x2], 'ko')  # Pontos em t=1s e 4s
    plot([0,5], [b, a*5+b])        # Secante
    plot([t1,t2], [x1, x1], '--')  # Delta t
    plot([t2,t2], [x1, x2], '--') # Delta x
    
    text(0.0, 60, """    x2-x1: %.2f m
    t2-t1: %.2f (s)
    v=%.2f m/s""" % (dx, dt, dx/dt))

    xlabel("t (s)")
    ylabel("x (m)")
    ylim(-5,100)
    grid()

interact(plot_secant, x=widgets.FloatSlider(min=0.01,max=4.0,step=0.01,value=3, description=r'$\Delta t$'));
```

#### Exemplo:

Calcular derivada de $x(t)=ct^3$, onde $c=$cte.

## Aceleração

- $a$ indica a variação da velocidade
- <div class="fragment">Acel. média no intervalo $[t_1, t_2]$:  $\bar{a} = $<span class="fragment">$\frac{v(t_2)-v(t_1)}{t_2-t_1}=\frac{\Delta v}{\Delta t}$</span></div>
- <div class="fragment">Acel. instantânea:  $a(t) = $<span class="fragment">$\lim_{\Delta t \rightarrow 0}\frac{v(t+\Delta t)-v(t)}{\Delta t} = \frac{dv}{dt}$</span></div>
- <div class="fragment">Unidade de $a$ (no SI): $\rightarrow$<span class="fragment">$\frac{\Delta v}{\Delta t}=\frac{m/s}{s}=\frac{m}{s^2}$</span></div>

Em termos de $x(t)$, $a(t) = \frac{dv}{dt} = \frac{d}{dt}\left(\frac{dx}{dt}\right) $<span class="fragment">$ = \frac{d^2x}{dt^2} $</span><span class="fragment">(segunda derivada da posição)</span>

### Caso particular: $a=a_0=$ cte
-Importante em casos como queda livre próximo a superfície terrestre, onde $g\approx 9,8\text{ m/s}^2$.
- <div class="fragment">Como obter $v$ partindo de $a$? E $x$ partindo do $v$ obtido?</div>

```python
figsize(4,3)
t = linspace(0,2,3)
a = ones(t.shape) * 1

plot(t, a)
fill_between(t[0:2],a[0:2],0,color='r',alpha=0.3)
xlabel('t (s)')
ylabel('a (m/s^2)')
xticks([])
ylim(0,1.2)
yticks([1],['a0'])
xticks([0,1], ['t0=0', 't'])
figsize(*default_figsize)
```

- Área sombreada: $a t = \Delta v$, ou <span class="fragment">$v=v_0+a t$.</span>

#### E $x(t)$?

```python
figsize(4,3)
t = linspace(0,2,3)
v = 1 + t

plot(t, v)
fill_between(t[0:2],v[0:2],0,color='r',alpha=0.3)
xlabel('t (s)')
ylabel('v (m/s)')
xticks([])
ylim(0,3)
yticks([1, 2],['v0', 'v(t)'])
xticks([0,1], ['t0=0', 't'])
hlines(2, 0, 1, linestyles='dashed')

figsize(*default_figsize)
```

- Área sombreada: $\Delta x = \frac{[v_0 + v(t)]t}{2}$ =<span class="fragment">
    $\frac{[v_0 + v_0+at]t}{2} =$</span><span class="fragment">
    $v_0 t + a \frac{t^2}{2}$</span>
    
    <span class="fragment">$x = x_0 + v_0 t + a \frac{t^2}{2}$</span>

- <div class="fragment">Obtivemos a equação da posição partindo diretamente da curva de aceleração!</div>
